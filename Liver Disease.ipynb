```python
def missing_values(data):
    total = data.isnull().sum().sort_values(ascending=False)
    percentage = (data.isnull().sum()/data.isnull().count()).sort_values(ascending=False)
    percentage_final = (round(percentage, 2) * 100)
    total_percent = pd.concat(objs=[total, percentage_final], axis = 1, keys=['Total', '%'])
    return total_percent
```

## 模型建立及训练

```python
import warnings
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
warnings.filterwarnings('ignore')
```

MinMaxScaler类用于对数据进行最小-最大缩放。它通过将数据按照最小值和最大值的范围进行线性缩放，将数据映射到指定的区间（默认为[0, 1]）。这种缩放方法可以保留原始数据的分布形状，并适用于对异常值不敏感的情况。

StandardScaler类用于对数据进行标准化缩放。它通过计算数据的均值和标准差，将数据进行线性变换，使得数据的均值为0，方差为1。这种缩放方法适用于对数据分布形状没有要求，并且能够将数据转换为近似标准正态分布的形式。

```python
pd.get_dummies(df_liver['Gender'], prefix = 'Gender').head()
```

```python
# 加入独热向量
df = pd.concat([df_liver,pd.get_dummies(df_liver['Gender'], prefix = 'Gender')], axis=1)
df.head()
```

```python
X1 = df_liver.drop(['Gender', 'Direct_Bilirubin','Alamine_Aminotransferase','Albumin','class'], axis=1)
X = df_liver.drop(['Gender', 'class'], axis=1)

y = df_liver['class']
```

df.drop(['Gender', 'class'], axis=1)将DataFrame对象df中的'Gender'和'class'列删除，并将结果存储在变量X中。axis=1表示按列删除，即删除指定的列。

df['class']选取了DataFrame对象df中的'class'列，并将结果存储在变量y中。这是目标变量，即模型需要预测或分类的目标。

### 归一化

```python
# 创建归一化器
scaler = StandardScaler()

# 归一化
X = scaler.fit_transform(X)
X1 = scaler.fit_transform(X1)
```

StandardScaler()创建了一个StandardScaler的实例，用于进行数据标准化缩放。

scaler.fit_transform(X)调用了StandardScaler对象的fit_transform()方法，该方法接受一个特征矩阵X作为输入。

fit_transform()方法首先使用X计算特征矩阵的均值和标准差，以便后续进行标准化操作。

然后，fit_transform()方法对特征矩阵X进行标准化缩放操作，将每个特征的值减去均值，然后除以标准差，使得每个特征的均值为0，标准差为1。

最后，标准化后的数据存储在变量X中，可以用于后续的数据分析或机器学习任务。

```python
# 划分训练集和测试集
X_train, X_test, y_train, y_test= train_test_split(X, y, train_size=0.8, random_state=42)
X1_train, X1_test, y_train, y_test= train_test_split(X1, y, train_size=0.8, random_state=42)
```

```python
model = SVC(C = 10, kernel = "sigmoid",probability=True)
model.fit(X_train, y_train)
```

```python
model1 = SVC(C = 10, kernel = "sigmoid",probability=True)
model1.fit(X1_train, y_train)
```

```python
y_pred = model.predict(X_test)
accuracy = model.score(X_test,y_test)
print('准确率',accuracy)
print('支持向量机模型准确率:',accuracy_score(y_test,y_pred))
print(classification_report(y_test, y_pred)) 
```
